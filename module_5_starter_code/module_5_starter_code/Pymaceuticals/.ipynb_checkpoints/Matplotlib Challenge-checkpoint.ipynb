# Analysis
Summary statistics table of the tumor volume for each drug regimen showed that:
(a) Drugs Ramicane had ~approx. similar values of mean and median values, which suggest that the data can be normally distributed for this drug
(a) Drugs Capomulin and Ramicane had low standard deviations (close to 0) as compared to the other drugs. A standard deviation close to zero indicates that data points are close to the mean, whereas a high or low standard deviation indicates data points are respectively above or below the mean.
(b) Similarly drugs Capomulin and Ramicane showed lower standard errors as compared to others. This suggests that the data accuracy is higher for population distribution.
Bar and Pie Charts suggest that:
(a) Number of mice tested for Drugs Capomulin and Ramicane are the highest as compared to other drugs.
(b) The Male sex (51%) of mice tested is more as compared to Female ones.
Quartiles, Outliers and Boxplots
(a) The outliers are present for Capomulin, Ramicane, Infubinol but its not present for Ceftamin
(b) Boxplot suggest that data representing drug Capomulin data is left skewed, Ramicane is slight left skewed, for Infubinol no skewedness is seen, Ceftamin also the data set is left skewed
R Value:0.71
A positive r-value coefficient indicates a positive relationship between average tumor volume vs. weight of mice

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import sklearn.datasets as dta
import scipy.stats as st
import matplotlib.pyplot as plt
from scipy.stats import sem
from scipy.stats import linregress
```

```python
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"
# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
mouse_metadata.head()
```

```python
# Combine the data into a single dataset
Combined_dataset = mouse_metadata.merge(study_results, on='Mouse ID')
# Display the data table for preview
Combined_dataset
```

```python
# Checking the number of mice.
Total_mice = len(Combined_dataset["Mouse ID"].value_counts())
Total_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
Duplicate_data_mouseID = Combined_dataset[Combined_dataset.duplicated(subset=['Mouse ID', 'Timepoint'])]['Mouse ID'].unique()
Duplicate_data_mouseID
```

```python
# # Optional: Get all the data for the duplicate mouse ID. 
Duplicate_data_mouseID = Combined_dataset[Combined_dataset.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]
Duplicate_data_mouseID
```

```python
#  Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = Combined_dataset.loc[Combined_dataset['Mouse ID'] != 'g989', :]
clean_df = clean_df.sort_values('Drug Regimen')
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
Total_mice_clean_df = len(clean_df["Mouse ID"].unique())
Total_mice_clean_df
```

# Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
Mean_Tumor_Volume = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
Mean_Tumor_Volume
Median_Tumor_Volume = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
Median_Tumor_Volume
Tumor_Volume_Variance = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
Tumor_Volume_Variance 
Tumor_Volume_Std_Dev = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
Tumor_Volume_Std_Dev
Tumor_Volume_Std_Err = clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()
Tumor_Volume_Std_Err

summary_dataframe = pd.DataFrame({"Mean Tumor Volume" : Mean_Tumor_Volume, "Median Tumor Volume" : Median_Tumor_Volume,
                    "Tumor Volume Variance" : Tumor_Volume_Variance, "Tumor Volume Std. Dev." : Tumor_Volume_Std_Dev, 
                    "Tumor Volume Std. Err." : Tumor_Volume_Std_Err})
summary_dataframe
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line.
summary_drug_al = clean_df.groupby('Drug Regimen').aggregate({'Tumor Volume (mm3)' : ['mean', 'median', 'var', 'std', 'sem']})
summary_drug_al
```

# Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse 
final_timepoint_s = clean_df.groupby(['Mouse ID'])['Timepoint'].max()
final_timepoint_s
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_timepoint_data_df = pd.merge(final_timepoint_s, clean_df, on=['Mouse ID', 'Timepoint'], how='left')
# # Display the data table for preview
final_timepoint_data_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
good_drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
# Create empty list to fill with tumor vol data
good_drugs_tumor_volumes = [] 
# for each drug 
for drug in good_drugs:
     # Locate the rows which contain mice on this drug and get the tumor volumes
    mice_on_drug_s = final_timepoint_data_df['Drug Regimen'] == drug
    drug_tumor_vol_s = final_timepoint_data_df[mice_on_drug_s]['Tumor Volume (mm3)'] 
    # Calculate the IQR and quantitatively determine if there are any potential outliers. for this drug
    drug_tumor_vol_s.quantile([0.25, 0.5, 0.75])
    qr3 = drug_tumor_vol_s.quantile([0.75])[0.75]
    qr1 = drug_tumor_vol_s.quantile([0.25])[0.25]
    iqr = qr3-qr1
    lowerbound = qr1-iqr
    upperbound = qr3+iqr
    print(f"{drug} IQR Calculation: IQR = {round(iqr, 2)}, lowerbound = {round(lowerbound, 2)}, upperbound = {round(upperbound, 2)}")
    potential_outlier_df = final_timepoint_data_df.loc[(final_timepoint_data_df['Tumor Volume (mm3)'] < lowerbound) | (final_timepoint_data_df['Tumor Volume (mm3)'] > upperbound), :]
# Outliers series
    potential_outliers_series = potential_outlier_df.loc[potential_outlier_df['Drug Regimen'] == drug, :]
    potential_outliers_series = potential_outliers_series["Tumor Volume (mm3)"]
    print(f"{drug} potential_outliers_series {round(potential_outliers_series, 2)}")
#   add subset the tumor vol data for this drug to the list of tumor vol data(for plotting)
    good_drugs_tumor_volumes.append(drug_tumor_vol_s.values)
```

```python
fig, ax = plt.subplots()
ax.boxplot(good_drugs_tumor_volumes)
ax.set_xticklabels(good_drugs)
ax.set_ylabel('Final Tumor Volume (mm3)')
plt.show()
```

# Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
sorted_drug_mouseID_df = clean_df.loc[(clean_df['Drug Regimen'] == 'Capomulin') & (clean_df['Mouse ID'] == 'l509'), :]
sorted_drug_mouseID_df = sorted_drug_mouseID_df.sort_values(['Timepoint'], ascending = True)
plt.plot(sorted_drug_mouseID_df['Timepoint'], sorted_drug_mouseID_df['Tumor Volume (mm3)'], color='blue')
plt.xlabel("Timepoint (days)")
plt.ylabel("Total Volume (mm3)")
plt.title("Capomulin treatment of mouse I509")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
Drug_sorted_df = clean_df.loc[(clean_df['Drug Regimen'] == 'Capomulin'), :]
average_tumor_volume_s = Drug_sorted_df.groupby(['Mouse ID', 'Weight (g)'])['Tumor Volume (mm3)'].mean()
average_tumor_volume_df = pd.DataFrame({"Average tumor volume (mm3)" : average_tumor_volume_s})
average_tumor_volume_df = average_tumor_volume_df.reset_index()
plt.scatter(average_tumor_volume_df['Weight (g)'], average_tumor_volume_df['Average tumor volume (mm3)'], marker='o', color='blue')
plt.xlabel("Weight (g)")
plt.ylabel("Average tumor volume (mm3)")
plt.show()
```

# Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
# Calculate the correlation coefficient between mouse weight and average tumor volume for the Capomulin regimen
print(st.pearsonr(average_tumor_volume_df['Weight (g)'], average_tumor_volume_df['Average tumor volume (mm3)']))
(slope, intercept, rvalue, pvalue, stderr) = linregress(average_tumor_volume_df['Weight (g)'], average_tumor_volume_df['Average tumor volume (mm3)'])
```

```python
# Get regression values
regress_values = slope * average_tumor_volume_df['Weight (g)'] + intercept
```

```python
line_eq = f"y = {round(slope, 2)}x + {round(intercept, 2)}"
print(line_eq)
```

```python
plt.scatter(average_tumor_volume_df['Weight (g)'], average_tumor_volume_df['Average tumor volume (mm3)'])
plt.xlabel("Weight (g)")
plt.ylabel("Average tumor volume (mm3)")
plt.plot(average_tumor_volume_df['Weight (g)'], regress_values, color='red')
# Label plot and annotate the line equation
plt.annotate(line_eq,(22,40),fontsize=10,color="red")
# Print r square value
print(f"The r-value is: {round(rvalue**2, 2)}")
# Show plot
plt.show()
```
